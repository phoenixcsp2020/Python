```python
# 5. Correlation between age and salary
cor_re = df['Age'].corr(df['Salary'])
cor_re
```

```python
# Visual representation of correlation
sns.scatterplot(data=df, x='Age',y='Salary', marker='*', edgecolor='b')
plt.title('Correlation between Age and Salary', color='r')
plt.xlabel("Age of the employee", color='m')
plt.ylabel("Salary of the EMP", color='b')
plt.figure(figsize=(7,5))
plt.show()
```
